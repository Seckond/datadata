```python
#https://stevenloria.com/tf-idf/

import math
from textblob import TextBlob as tb
import pandas as pd
import numpy as np
```

```python
def tf(word, blob):
    return blob.words.count(word) / len(blob.words)

def n_containing(word, bloblist):
    return sum(1 for blob in bloblist if word in blob)

def idf(word, bloblist):
    return math.log(len(bloblist) / (1 + n_containing(word, bloblist)))

def tfidf(word, blob, bloblist, idf_boost=0):
    return (tf(word, blob)**(1 - idf_boost)) * (idf(word, bloblist)**(1 + idf_boost))
```

```python
drexler_df = pd.read_csv('data/drexler_complete_lyrics.csv', sep=';')
drexler_df['words'] = drexler_df.lyrics.str.strip().str.split('[\W_]+')
drexler_df.sample(10)
```

```python
related_df = pd.read_csv('data/related_lyrics.csv', sep=';')
related_df['words'] = related_df.lyrics.str.strip().str.split('[\W_]+')
related_df.sample(10)
```

```python
for i in range(len(drexler_df)):
    print("%d: %.10f"% (i, tf('mar',tb(drexler_df['lyrics'].loc[i]))))
```

```python
from nltk.corpus import stopwords
stopwords = stopwords.words('spanish')
```

```python
drexler_corpora = [tb(drexler_df['lyrics'].loc[i].lower()) for i in range(len(drexler_df))]
related_corpora = [tb(related_df['lyrics'].loc[i].lower()) for i in range(len(related_df))]
```

```python


for i, blob in enumerate(drexler_corpora):
    print("Top words in document %s" % drexler_df['title'].loc[i])    
    scores = {word: tfidf(word, blob, related_corpora, 0.7) for word in blob.words if (word not in stopwords)}
    sorted_words = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    for word, score in sorted_words[:5]:
        print("\tWord: {}, TF-IDF: {}".format(word, round(score, 5)))
  
    
    
    
```
